Today, I'm looking at ChatGPT's ability to do understand the structure of Census data. 
The [data](https://www.census.gov/library/publications/1953/dec/housing-vol-02.html) are organized into different PDFs as follows. 

* Part 1: United States and Geographic Divisions 

* Part 2: Akron, OH  Des Moines, IA

* Part 3: Detroit, MI  Memphis, TN

* Part 4: Miami, FL  Salt Lake City, UT

* Part 5: San Antonio, TX  Youngstown, OH

### The task 

I was curious if ChatGPT would be able to tell me which section would contain data for Philadelphia. 



Inspired by the concept of *prewarming* or *internal retrieval* described in *Prompt Engineering for Generative AI* (Phoenix and Taylor 2024), I decided to ask ChatGPT if it could find a pattern in the file structures. Then, referencing our previous exchange where it discovered the pattern, I would ask it to try the task again. 




Ok, time for the internal retrieval to shine. 



<div class="alert alert-block alert-success">
<b>Pros:</b> ChatGPT was able to remember that it had previously found that the files were ordered alphabetically when I started the next prompt with "With that in mind". 
</div>

<div class="alert alert-block alert-danger">
<b>Cons: </b> ChatGPT is not correctly retrieving the order of the alphabet.
</div>

After expressing some doubt in the alphabetization, ChatGPT was able to provide me with the correct answer.



And just to be sure, I asked it about a different US city, and it was able to complete the task.




### So, could this have been done all in one prompt?

I closed out my current session of ChatGPT and tried to create one prompt that would make ChatGPT do some sort of internal retrieval. 

I ended up trying this: 

> Hi! I have a two part question. These are pdf names that have housing data for various US cities in 1950...

> **Question 1**: Why do you think Akron, OH comes before Des Moines, IA, and Des Moines, IA comes before Detroit, MI? Explain your reasoning.
> **Question 2**: Which pdf do you think Philadelphia, PA would be in? Explain your reasoning.

ChatGPT was able to correctly answer **Question 1**: saying "The cities are likely listed in alphabetical order (from A to Z), which is why Akron, OH appears before Des Moines, IA, and Des Moines, IA comes before Detroit, MI". 

But when it came to **Question 2**, there was more work still to be done.



Then I thought maybe a three part question would help: 

> **Question 1**: Why do you think Akron, OH comes before Des Moines, IA, and Des Moines, IA comes before Detroit, MI? Explain your reasoning.
> **Question 2**: Can you please print out the order of the alphabet? 
> **Question 3**: Which pdf do you think Philadelphia, PA would be in? Explain your reasoning.
