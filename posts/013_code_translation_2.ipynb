
It was pretty good at working with R, and nowhere near as good at working with Stata, which became really apparent in steps 3 and 4. When I asked it to translate back to Stata, it gave me some syntax from MATLAB! 

I was motivated to do this exercise because I am by far the most comfortable working in Stata, but Stata is not widely used outside of the field of economics. I want to increase my competency in other languages so that my skillset is more versatile, and I figured ChatGPT might be able to support this mission. But since Stata is pretty niche, ChatGPT was not able to produce Stata code very well. 

### Today's Mission ###
Today, I'm going to take the tricky part of the experiment, doing steps 3 and 4, to Chatbot Area. This will produce several anonymized LLM responses, and I will keep voting until I see which LLM did the best. 

I want to craft a prompt that really tests the task that ChatGPT was not able to do well last time around. So, I need to provide some R code and ask for a translation to Stata. 

In the spirit of my original post on this topic, I am going to get some R code from a separate window of ChatGPT. My prompt is: 

> I want you to create simple, easy to understand R code that draws a random sample of 100 numbers from a normal distribution centered at 0 with a standard deviation of 1. Then, the code will create a simple histogram displaying the results. The x axis will show the values drawn. The y axis will show the frequency with which each value was drawn. Key things to remember: #1: easy to read, #2: simple, #3, creates proper frequency histogram.

I got back the following code: 

> ##### Set seed for reproducibility
> set.seed(123)
> 
> ##### Draw 100 random numbers from a normal distribution (mean = 0, sd = 1)
> sample_data <- rnorm(100, mean = 0, sd = 1)
> 
> ##### Create a histogram
> hist(sample_data, 
     main = "Histogram of Random Normal Sample", 
     xlab = "Value", 
     ylab = "Frequency", 
     col = "lightblue", 
     border = "black")

Just to be sure, I confirmed that it produces the desired graph in R: 



### Time for the Arena! ###
Now I'm going to provide this R code and ask LLMs to translate it to Stata. My prompt will be this: 

> You are a helpful assistant skilled in writing code in Stata. Please translate the following R code to Stata.
>  
> set.seed(123)
> 
> sample_data <- rnorm(100, mean = 0, sd = 1)
> 
> hist(sample_data, 
     main = "Histogram of Random Normal Sample", 
     xlab = "Value", 
     ylab = "Frequency", 
     col = "lightblue", 
     border = "black")

I've removed the comments that ChatGPT had produced in an attempt to make this a little harder. 

Now I'm going through, testing the Stata output code. Here are some of the errors I got along the process: 





#### Some models that did kind of work ####

This one, made by cybele, has density and not frequency on the y-axis, but at least it runs. 



Density and not frequency on the y-axis once again.



And again! 




#### Conclusion #### 
It seems like we are still far away from LLMs being able to write perfect Stata code, but their errors are minor on the most part and easy to troubleshoot. 

There were some remarkable similarities across all of these models: 

1. The data generated was called `sample_data`

2. The graph title was "Histogram of Random Normal Sample"

3. The y-axis title was "Frequency"

4. The x-axis title was "Value"

5. The seed set was 123

That being said, these all mimic the labels and titles in the prompt I provided! I wonder if a different time, if I try asking for the code directly across different models, if the titles, data name, and seed number would still be so consistent. I'll have to try that some other time. 

Overall, this time around, `chatgpt-4o-latest-20250326` was the winner- a surprising comeback given that this post was inspired by ChatGPT not doing well with these tasks in the first place.
